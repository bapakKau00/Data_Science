```python
import pandas as pd

data = pd.read_csv("housing.csv")

data.head(10)
```

```python
data.describe()
```

```python
from statistics import mean, median, mode, stdev
#dataCol = ['housing_median_age', 'total_rooms', 'median_house_value'] 

mean = data['housing_median_age'].mean()
print("Mean for housing_median age : " , mean)


median = data['housing_median_age'].median()
print('Median for housing_median age :' , median)


stdDev = data['housing_median_age'].std()
print('Standard deviation for housing_median age :' , stdDev)

mode = data['housing_median_age'].mode()
print('\nMode for housing_median age :\n' , mode)

```

```python
mean = data['total_rooms'].mean()
print("Mean for total_rooms : " , mean)


median = data['total_rooms'].median()
print('Median for total_rooms :' , median)


stdDev = data['total_rooms'].std()
print('Standard deviation for total_rooms :' , stdDev)

mode = data['total_rooms'].mode()
print('\nMode for total_rooms:\n' , mode)

```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(8, 6))
plt.hist(data['total_rooms'], bins=20, alpha=0.7)
plt.axvline(data['total_rooms'].mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(data['total_rooms'].median(), color='green', linestyle='dashed', linewidth=2, label='Median')
plt.legend()
plt.title('Histogram with Mean and Median of total_rooms')
plt.show()
```

```python
mean = data['population'].mean()
print("Mean for population : " , mean)

median = data['population'].median()
print('Median for population :' , median)

stdDev = data['population'].std()
print('Standard deviation for population :' , stdDev)

mode = data['population'].mode()
print('\nMode for population:\n' , mode)

```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(8, 6))
plt.hist(data['population'], bins=20, alpha=0.7)
plt.axvline(data['population'].mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(data['population'].median(), color='green', linestyle='dashed', linewidth=2, label='Median')
plt.legend()
plt.title('Histogram with Mean and Median of population')
plt.show()
```
